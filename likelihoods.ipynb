```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
infile = 'example_data/wsdi_yr_ACCESS-CM2_ssp245_ensemble_aus-states-cities_1850-2100.csv'
```

```python
df = pd.read_csv(infile)
```

```python
df
```

```python
location = 'AUS'
ref_start = 1950
ref_end = 2014

df_ref = df[(df['year'] >= ref_start) & (df['year'] <= ref_end)]
nruns = len(df['run'].unique())
window = nruns * 20

df_data = df.drop(['year', 'model', 'run', 'experiment'], axis=1)
df_ref_data = df_ref.drop(['year', 'model', 'run', 'experiment'], axis=1)

p98 = df_ref_data.quantile(0.98)
```

```python
p98
```

```python
df_threshold_test = df_data > p98
```

```python
odds = df_threshold_test.rolling(window, step=nruns, center=True).apply(lambda s: (sum(s) / window) * 100)
odds.index = df['year'].unique()
odds = odds.dropna()
odds.index.name = 'year'
odds = odds.round(decimals=1)
```

```python
odds
```

```python
model = df['model'].unique()[0]
```

```python
experiment = df['experiment'].unique()[-1]
```

```python
runs = list(df['run'].unique())
```

```python
f"{runs}"
```

```python
pre_header_lines = [
    "Metric: WSDI",
    f"Model: {model}",
    f"Experiment: {experiment}",
    f"Runs: {runs}",
    "Command: <insert cmdline_provenance>"
]

output_filename = 'output_with_pre_header.csv'

# Open the file and write the pre-header lines
with open(output_filename, mode='w') as f:
    for line in pre_header_lines:
        f.write(line + '\n')
    odds.to_csv(f, mode='a', header=True, index=True)
```
